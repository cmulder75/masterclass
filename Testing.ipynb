```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Hyper-parameters for the Scraper
date_from = '2021-11-20'
date_to = '2021-12-04'
tweet_limit = 200          # The corpus need to be between 500 - 1000 entries
virus_keyword = 'omicron'

# Create list to append tweet data to
tweets_list = []

# Create the Search Term
search_term = virus_keyword + ' since:' + date_from + ' until:' + date_to + ' lang:en'
print(search_term)
```

```python
# Using TwitterSearchScraper to scrape data and append tweets to list
twitter_items = sntwitter.TwitterSearchScraper(search_term).get_items()

for i,tweet in enumerate(twitter_items):
    # Only retrieve the tweet_limit of tweets
    if i>tweet_limit:
        break

    tweet_length = len(str(tweet.content))
    print(dir(tweet))
    
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.username, tweet_length])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Length'])
display(tweets_df)
```

```python
for index, row in tweets_df.iterrows():
    print(len(row['Text']))
```

```python
from datetime import datetime, timedelta

def validate_date(input_date):
        try:
            # Attempt to construct a date from the input_date
            newDate = datetime.strptime(input_date, "%Y-%m-%d")
            return newDate
        except ValueError:
            return None
```

```python
def calc_from_date(input_datetime, number_of_days):
    try:
        from_datetime = input_datetime - timedelta(days=number_of_days)
        return from_datetime 
    except ValueError:
        return None
    except TypeError:
        return None
```
